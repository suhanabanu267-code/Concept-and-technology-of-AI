3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
Based on the dataset Answer the following question:
What percent of the total fare revenue came from each passenger class?
To answer the question perform following operation:
1. Calculate the total Fare paid across all passengers in the Titanic dataset.

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
total_fare = titanic['Fare'].sum()
print("Total Fare paid by all passengers:", total_fare)
```

2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class_passengers = titanic[titanic['Pclass'] == 1]
total_fare_first_class = first_class_passengers['Fare'].sum()
print("Total Fare paid by first-class passengers:", total_fare_first_class)
```

3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
total_fare_second_class = titanic[titanic['Pclass'] == 2]['Fare'].sum()
print("Total Fare paid by second-class passengers:", total_fare_second_class)

total_fare_third_class = titanic[titanic['Pclass'] == 3]['Fare'].sum()
print("Total Fare paid by third-class passengers:", total_fare_third_class)
```

4. Combine the fare totals from first, second, and third classes into a list.

```python
total_fare_first = titanic[titanic['Pclass'] == 1]['Fare'].sum()
total_fare_second = titanic[titanic['Pclass'] == 2]['Fare'].sum()
total_fare_third = titanic[titanic['Pclass'] == 3]['Fare'].sum()

fare_totals = [total_fare_first, total_fare_second, total_fare_third]

print("Fare totals by class [1st, 2nd, 3rd]:", fare_totals)
```

5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
total_fare_first = titanic[titanic['Pclass'] == 1]['Fare'].sum()
total_fare_second = titanic[titanic['Pclass'] == 2]['Fare'].sum()
total_fare_third = titanic[titanic['Pclass'] == 3]['Fare'].sum()

overall_total_fare = titanic['Fare'].sum()

fare_proportions = [
    total_fare_first / overall_total_fare,
    total_fare_second / overall_total_fare,
    total_fare_third / overall_total_fare
]

print("Proportion of fare revenue by class [1st, 2nd, 3rd]:", fare_proportions)
```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?
To answer the question perform following operation:
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

titanic['AgeGroup'] = titanic['Age'].apply(categorize_age)

print(titanic[['Age', 'AgeGroup']].head())
```

2. Calculate the total number of passengers on the Titanic.

```python
total_passengers = len(titanic)

print("Total number of passengers on the Titanic:", total_passengers)
```

3. Count the number of passengers in each age group.

```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

titanic['AgeGroup'] = titanic['Age'].apply(categorize_age)

age_group_counts = titanic['AgeGroup'].value_counts()

print("Number of passengers in each age group:")
print(age_group_counts)
```

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

titanic['AgeGroup'] = titanic['Age'].apply(categorize_age)

age_group_counts = titanic['AgeGroup'].value_counts()

total_passengers = len(titanic)

age_group_proportions = age_group_counts / total_passengers

print("Proportion of passengers in each age group:")
print(age_group_proportions)
```

5. Display the proportion as a percentage.

```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

titanic['AgeGroup'] = titanic['Age'].apply(categorize_age)

age_group_counts = titanic['AgeGroup'].value_counts()

total_passengers = len(titanic)

age_group_percent = (age_group_counts / total_passengers) * 100

print("Percentage of passengers in each age group:")
print(age_group_percent.round(2))
```
