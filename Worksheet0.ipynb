```python
#Task 1 - Sum of Nested Lists:


def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)  
        else:
            total += element 
    return total


nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list)) 
```

```python
#Task 2 - Generate All Permutations of a String:

def generate_permutations(s):
    if len(s) == 1:
        return [s] 
    
    permutations = set()  
    
    for i, char in enumerate(s):
       
        remaining = s[:i] + s[i+1:]
     
        for perm in generate_permutations(remaining):
            permutations.add(char + perm) 
    
    return list(permutations)

# Test cases
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
#Task 3 - Directory Size Calculation

def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # Subdirectory
            total_size += calculate_directory_size(value)
        else:  # File size
            total_size += value
    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
print(calculate_directory_size(directory_structure))  # Expected output: 1400
```

```python
#Task 1 - Coin Change Problem


def min_coins(coins, amount):
    # Initialize DP array with infinity
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed to make amount 0

    # Build up dp array
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test
coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))  # Expected output: 3
```

```python
#Task 2 - Longest Common Subsequence (LCS)

def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]  # Create DP table

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

# Test
s1 = "abcde"
s2 = "ace"
print(longest_common_subsequence(s1, s2))  # Expected output: 3
```

```python
#Task 3 - 0/1 Knapsack Problem

def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] > w:
                dp[i][w] = dp[i - 1][w]
            else:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])

    return dp[n][capacity]

# Test
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))  # Expected output: 9
```

```python
#Task 1. Classify Study Time:


time_data = [2, 4, 6, 1.5, 3, 5.5, 4.5, 2.5]
low = []
moderate = []
high = []


for hours in time_data:
    if hours < 3:
        low.append(hours)
    elif 3 <= hours <= 5:
        moderate.append(hours)
    else:  # hours > 5
        high.append(hours)


print("Low study time (<3 hours):", low)
print("Moderate study time (3-5 hours):", moderate)
print("High study time (>5 hours):", high)
```

```python
#Task 2. Based on Data

low_days = len(low)
print("Number of days with low study time:", low_days)

moderate_days = len(moderate)
print("Number of days with moderate study time:", moderate_days)

high_days = len(high)
print("Number of days with high study time:", high_days)
```

```python
#Task 4. Analyze Average Time Use:

time_data = [
    [2, 3, 7],
    [4, 2, 6.5],
    [6, 1, 8],
    [1.5, 4, 6],
    [3, 2.5, 7],
    [5.5, 1.5, 6.5],
    [4.5, 2, 7.5],
    [2.5, 3.5, 6]
]

study_hours = []
entertainment_hours = []
sleep_hours = []

for day in time_data:
    study_hours.append(day[0])
    entertainment_hours.append(day[1])
    sleep_hours.append(day[2])

avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)

print(f"Average hours spent studying: {avg_study:.2f}")
print(f"Average hours spent on entertainment: {avg_entertainment:.2f}")
print(f"Average hours spent sleeping: {avg_sleep:.2f}")
```
