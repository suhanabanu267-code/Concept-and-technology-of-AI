```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Houseprice (1).csv')
df.head()
```

```python
house_age = df['HouseAge'].to_numpy()
house_floor = df['HouseFloor'].to_numpy()
house_area = df['HouseArea'].to_numpy()
house_price = df['HousePrice'].to_numpy()
```

```python
x0 = np.ones(len(house_age))
X = np.array([x0, house_age, house_floor, house_area]).T
W = np.array([0, 0, 0, 0])
Y = np.array(house_price)
```

```python
def cost_function(X, Y, W):
    m = len(Y)
    J = np.sum((X.dot(W) - Y) ** 2) / (2 * m)
    return J
```

```python
X_test = np.array([[1, 2], [3, 4], [5, 6]])
Y_test = np.array([3, 7, 11])
W_test = np.array([1, 1])
cost = cost_function(X_test, Y_test, W_test)

if cost == 0:
    print("Proceed Further")
else:
    print("Cost function output:", cost)
```

```python
initial_cost = cost_function(X, Y, W)
print("Initial cost:", initial_cost)
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)

    for i in range(iterations):
        Y_pred = X.dot(W)
        loss = Y_pred - Y
        dw = (X.T.dot(loss)) / m
        W = W - alpha * dw
        cost_history[i] = cost_function(X, Y, W)

    return W, cost_history
```

```python
alpha = 0.000001
iterations = 100000
new_weights, cost_history = gradient_descent(X, Y, W, alpha, iterations)

print("Trained weights:", new_weights)
print("Final cost:", cost_history[-1])
```

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))

def r2(Y, Y_pred):
    ss_tot = np.sum((Y - np.mean(Y)) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
Y_pred = X.dot(new_weights)
print("RMSE:", rmse(Y, Y_pred))
print("R²:", r2(Y, Y_pred))
```
